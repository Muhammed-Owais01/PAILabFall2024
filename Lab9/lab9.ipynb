```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Iris.csv')
```

### Task 1

```python
sepal_length = df['SepalLengthCm']
petal_length = df['PetalLengthCm']

plt.figure(figsize=(10, 6))
plt.plot(sepal_length, petal_length, marker='o', linestyle='-', color='b', markersize=5)

plt.title('Petal Length vs Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.grid()

# Show plot
plt.show()
```

### Task 2

```python
species_counts = df['Species'].value_counts()

plt.figure(figsize=(8, 5))
species_counts.plot(kind='bar', color=['blue', 'orange', 'green'])

plt.title('Number of Different Species in the Iris Dataset')
plt.xlabel('Species')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y')

plt.tight_layout()
plt.show()
```

### Task 3

```python
plt.figure(figsize=(10, 6))

colors = {'Iris-setosa': 'red', 'Iris-versicolor': 'blue', 'Iris-virginica': 'green'}
plt.scatter(df['Species'], df['PetalWidthCm'], 
            c=df['Species'].map(colors), 
            alpha=0.6, edgecolors='w', s=100)

plt.title('Scatter Plot of Species vs Petal Width')
plt.xlabel('Species')
plt.ylabel('Petal Width (cm)')
plt.grid()

plt.tight_layout()
plt.show()
```
