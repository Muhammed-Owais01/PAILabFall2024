```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Q1

```python
student_names = [f"Students {i + 1}" for i in range(20)]
student_heights = [150, 160, 155, 158, 165, 170, 175, 160, 180, 168,
                   158, 162, 170, 155, 167, 159, 173, 169, 164, 172]

colors = plt.cm.viridis(np.linspace(0, 1, 20))

plt.figure(figsize=(14, 6))


plt.subplot(1, 2, 1)

# Bar chart
plt.bar(student_names, student_heights, color=colors)
plt.title("Student Data")
plt.xlabel("Name")
plt.ylabel("Height")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()

plt.subplot(1, 2, 2)
plt.pie(student_heights, labels=student_names, colors=colors, autopct="%1.1f%%")
plt.title("Student Data")
plt.show()
```

# Q4

```python
df = pd.read_csv("student-dataset.csv")

age_groups = ['18-25', '26-30', '31-35', '36 and above']
group_counts = [0, 0, 0, 0]

for age in df["age"]:
    if 18 <= age <= 25:
        group_counts[0] += 1
    elif 26 <= age <= 30:
        group_counts[1] += 1
    elif 31 <= age <= 35:
        group_counts[2] += 1
    elif age >= 36:
        group_counts[3] += 1

plt.pie(group_counts, labels=age_groups, autopct="%1.1f%%")
plt.show()
```

# Q5

```python
labels = ["Male", "Female"]
gender_count = [0, 0]

for gender in df['gender']:
    if gender == "M":
        gender_count[0] += 1
    elif gender == "F":
        gender_count[1] += 1

plt.pie(gender_count, labels=labels, autopct='%1.1f%%')
plt.show()
```

# Q6

```python
math_marks = [85, 78, 92, 88, 76, 95, 89, 74, 91, 82]
science_marks = [80, 85, 91, 87, 79, 94, 88, 73, 90, 84]
student_ids = [f"Student {i+1}" for i in range(10)]

plt.figure(figsize=(8, 6))
plt.scatter(math_marks, science_marks, c='blue', marker='o', label='Students')
plt.title("Student Maths and Science Marks")
plt.xlabel("Math Marks")
plt.ylabel("Science Marks")

plt.legend(title="Legend")

for i, txt in enumerate(student_ids):
    plt.annotate(txt, (math_marks[i], science_marks[i]), textcoords='offset points', xytext=(5,5), ha='center')

plt.grid(True)
plt.tight_layout()
plt.show()
```
